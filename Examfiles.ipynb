```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('traindata.csv')
df.head()
```

```python
df1 = pd.read_csv('testData.csv')
df1.head()
```

```python
## Counting the missing values in train dataset

df.isna()
```

```python
## Joining the datasets

df2 = df.append(df1)
df2
```

```python
## Converting into numeric form depending on the column


df2['Gender'] = df2['Gender'].replace({'Male':'0','Female':'1'})
df2['Dependents'] = df2['Dependents'].replace({'3+':'3'})
df2['Married'] = df2['Married'].replace({'Yes':'1','No':'0'})
df2['Education'] = df2['Education'].replace({'Graduate':'1','Not Graduate':'0'})
df2['Self_Employed'] = df2['Self_Employed'].replace({'Yes':'1','No':'0'})
df2['Property_Area'] = df2['Property_Area'].replace({'Urban':'1','Rural':'0','Semiurban':'2'})
df2
```

```python
## filling the missing values

df2.apply(lambda x: np.where(x.isnull(), x.dropna().sample(len(x), replace=True), x))
```

```python
df2.dropna(inplace=True)
df2
```

```python
## Applying Logistic Regression

import pandas as pd
```

```python
X = df2[['ApplicantIncome','CoapplicantIncome','Credit_History','Property_Area']]
X
```

```python
y = df2[['Self_Employed']]
y
```

```python
## Filling missing values

import random as rnd

df3 = df2.copy()
df3['LoanAmount'].fillna((df3['LoanAmount'].median()),inplace=True)
df3['Dependents'].fillna((df3['Dependents'].median()),inplace=True)
df3['Loan_Amount_Term'].fillna((df3['Loan_Amount_Term'].mean()),inplace=True)
df3.isna().sum()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.30,random_state=1)

from sklearn.linear_model import LogisticRegression
logmodel=LogisticRegression()
logmodel.fit(X_train,y_train)
logmodel.predict([[3748,1668.0,1.0,2]])
```

```python
print('Training score is',logmodel.score(X_train,y_train)*100)
print('Testing score is',logmodel.score(X_test,y_test)*100)
```

```python
## Calculating Accuracy score

from sklearn.metrics import accuracy_score

ypred = logmodel.predict(X_test)
accscore =accuracy_score(y_test,ypred)*100
print('Accuracy score is',accscore)
```
